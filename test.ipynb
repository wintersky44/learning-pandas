### imports

```python
%matplotlib widget
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### datasets

```python
df = pd.read_csv("data/melb_data.csv")

job_market = pd.read_csv("data/india_job_market_2024_2026.csv")

food_data = pd.read_csv("data/foods_dataset.csv")
```

### testing

```python
grouped = job_market.groupby(['Job_Title'])

dataset = grouped.Salary_LPA.mean()

dataset = dataset.sort_values()

plt.figure(figsize=(12, 4))
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.title("Mean salary for every computer science role")
plt.ylim(15, 30)
plt.ylabel("Average Salary per year (in Lakhs)")
plt.subplots_adjust(bottom=0.41)
sns.barplot(data = dataset)
```
